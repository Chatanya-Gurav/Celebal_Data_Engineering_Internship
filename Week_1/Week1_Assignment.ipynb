```python
# Import libraries required for data manipulation and cleaning
import pandas as pd
import numpy as np
import re

# Display all columns while exploring the dataset
pd.set_option("display.max_columns", None)

# Increase column width for better readability
pd.set_option("display.max_colwidth", 100)
```

```python
# Read the CSV file into a Pandas DataFrame
df = pd.read_csv("Combined_dataset.csv")

# Display the number of rows and columns
print(f"Dataset loaded successfully.")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
# Display the first five records
display(df.head())

# Display the last five records
display(df.tail())
```

```python
# View the dimensions of the dataset
print("Dataset Shape:", df.shape)

# Display all column names
print("\nColumn Names:")
print(df.columns.tolist())
```

```python
# Display column names, data types, and non-null values
df.info()
```

```python
# Generate summary statistics for numerical and categorical columns
display(df.describe(include="all").T)
```

```python
# Calculate missing values and their percentage
missing_values = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum() / len(df)) * 100
})

# Display only columns with missing values
missing_values = missing_values[missing_values["Missing Values"] > 0]

# Sort columns by highest percentage of missing values
missing_values = missing_values.sort_values(
    by="Percentage",
    ascending=False
)

display(missing_values)
```

```python
# Separate columns based on data type
text_columns = df.select_dtypes(include="object").columns
numeric_columns = df.select_dtypes(include=["int64", "float64"]).columns

# Replace missing text values with a placeholder
df[text_columns] = df[text_columns].fillna("Not Available")

# Replace missing numeric values with zero
df[numeric_columns] = df[numeric_columns].fillna(0)

# Verify that all missing values have been handled
print("Remaining missing values:", df.isnull().sum().sum())
```

```python
# Count duplicate rows before removal
duplicate_count = df.duplicated().sum()

print("Duplicate rows before removal:", duplicate_count)

# Remove duplicate records
df = df.drop_duplicates()

print("Dataset shape after removing duplicates:", df.shape)
```

```python
# Select products with ratings greater than or equal to 4
high_rated_products = df[df["rating"] >= 4.0]

display(high_rated_products.head())
```

```python
# Create a subset containing relevant columns
selected_columns = df[
    ["title", "category", "final_price", "rating"]
]

display(selected_columns.head())
```

```python
#Convert Price to Numeric Format

# Function to remove currency symbols and commas from price values
def clean_price(price):

    # Return 0 for missing values
    if pd.isna(price):
        return 0.0

    # Keep only digits and decimal points
    cleaned = re.sub(r"[^0-9.]", "", str(price))

    return float(cleaned) if cleaned else 0.0


# Create a numeric version of the price column
df["final_price_numeric"] = df["final_price"].apply(clean_price)
```

```python
# The dataset does not contain a quantity column.
# ratings_count is used as a proxy variable.

df["total_amount"] = (
    df["final_price_numeric"] * df["ratings_count"]
)

display(
    df[
        [
            "final_price",
            "final_price_numeric",
            "ratings_count",
            "total_amount"
        ]
    ].head()
)
```

```python
# Define output file name
output_file = "Combined_dataset_cleaned.csv"

# Save the cleaned dataset without row indices
df.to_csv(output_file, index=False)

print(f"Cleaned dataset saved as '{output_file}'")
```
